## Creating Spending Flag

```python
### Aggregating Data with transform() Function
ords_prods_merge['avg_price'] =ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
### checking the columns "user_id" and the new created "avg_price"
ords_prods_merge[['user_id', 'avg_price']].head(10)
```

```python
### Creating spending flag for customers
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High_spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low_spender'
```

```python
### checking the new created column
ords_prods_merge[['user_id', 'avg_price', 'spending_flag']].head(10)
```

```python
# Examining Value Count on 'spending_flag' Column
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Creating Order Frequency Flag

```python
### Aggregating Data with transform() Function
ords_prods_merge['median_days'] =ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
### checking the columns "user_id" and the new created "median_days"
ords_prods_merge[['user_id', 'median_days']].head(10)
```

```python
### Creating Frequency flag for customers
ords_prods_merge.loc[ords_prods_merge['median_days'] >20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days'] <= 20) & (ords_prods_merge['median_days'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days'] <=10, 'frequency_flag'] = 'Frequent customer'
```
